```python
import tkinter as tk
from tkinter import *
from tkinter import ttk
from tkinter.scrolledtext import *

# NLP Packages
from textblob import TextBlob
import spacy
 
 # Structure and Layout
window=Tk()
window.title("Part of speech Tagger")
window.geometry("700x400")
window.config(background='black')

# TAB LAYOUT
tab_control=ttk.Notebook(window)
 
tab1=ttk.Frame(tab_control)
tab2=ttk.Frame(tab_control)
tab3=ttk.Frame(tab_control)

# ADD TABS TO NOTEBOOK

tab_control.add(tab2, text='File Processer')


 
label2=Label(tab2, text= 'File Processing',padx=5, pady=5)
label2.grid(column=0, row=0)



tab_control.pack(expand=1, fill='both')

about_label=Label(tab3,text="NLPiffy GUI V.0.0.1 \n Jesus saves @JCharisTech",pady=5,padx=5)
about_label.grid(column=0,row=1)

# Functions FOR NLP  FOR TAB ONE
def openfiles():
	file1=tk.filedialog.askopenfilename(filetypes=(("Text Files",".txt"),("All files","*")))
	read_text=open(file1,encoding='utf-8').read()
	displayed_file.insert(tk.END,read_text)


def get_tokens():
	raw_text = str(raw_entry.get())
	new_text = TextBlob(raw_text)
	final_text = list(str(new_text.words).split(" "))
	result = '\nTokens:{}'.format(final_text)
	tab1_display.insert(tk.END,result)


def clear_display_result():
	tab1_display.delete('1.0',END)

# Clear Result of Functions
def clear_result():
	tab2_display_text.delete('1.0',END)

# Functions for TAB 2 FILE PROCESSER
# Open File to Read and Process


def get_file_tokens():
	raw_text=displayed_file.get('1.0',tk.END)
	new_text=TextBlob(raw_text)
	final_text=list(str(new_text.words).split(" "))
	result='\nTokens:{}'.format(final_text)
	tab2_display_text.insert(tk.END,result)


def get_file_pos_tags():
    raw_text=displayed_file.get('1.0',tk.END)
    new_text=TextBlob(raw_text)
    final_list=list()
    for word in new_text.split():
        if word in ('ثمّ','لكي','ليست','أنها','اما','لن','حيث','الى','في','و','من','لم','ف','منها','له','أو','كل','هنا','فيه','لا','أيها','بأن','أنّ','إذن','أنّه','أن','ما','عن','على','ب','إن','لأن','أهمها','إلا','إلا','بعد','مما','قد','اليها','هي','إذ','منه','ان','او','لكن','فيها','ك','حين','حين','حول','لها','عدم','به','الا','اي','لذا') or word[0:]=='ها' or word[0:]=='ما':
             final_list.append(tuple([word,"P"]))
        elif word[-1]=='َ' or (len(word)==3 and word[-1]=='َ' or word[-1]=='ُ') or word[0]=='ي' or word[0]=='ت' or (word.startswith('ب') and len(word)>5)or (word.startswith('ا') and word.endswith('َ')) :
                final_list.append(tuple([word,"V"]))
        elif word.endswith('ون'):
            final_list.append(tuple([word,"V"]))
        elif word.endswith('ين'):
            final_list.append(tuple([word,"V"]))
        elif word[0:2]=='ال' or (word[-1:len(word)]=='ا' and len(word)>2)or word[-1]=='ً'or word[0]=='أ'or word[-1]=='ة':
              final_list.append(tuple([word,"N"]))
        elif len(word)==3  and not(word.endswith('ُ')) or word[-1]=='ٌ':
             final_list.append(tuple([word,"N"]))
        elif len(word)>=5 and  word[2]=='ا' or (word.endswith('ُ')):
              final_list.append(tuple([word,"N"]))
        elif word in('مشكلو','للذين','معظم','عجلون','معان','الكرك','جنوبي','بقاع','مرض','هو','هم','كبار','كيلوغرامات','آلاف','مؤذيين','للعالم','مشكلات','مستوى'):
            final_list.append(tuple([word,"N"]))
        elif (word.endswith('ِ') and (word.startswith('ب') or  word.startswith('ل'))):
             final_list.append(tuple([word,"N"]))
        elif  word[-1]=='ه'  or len(word)==[3] or word[0]=='أ' or word.endswith('ِ'):
            final_list.append(tuple([word,"N"]))
        elif  word.endswith('ات'):
             final_list.append(tuple([word,"N"]))
        elif  word.endswith('تن'):
             final_list.append(tuple([word,"N"]))
        elif  word.endswith('ان'):
             final_list.append(tuple([word,"N"]))
        elif  word.endswith('اء'):
             final_list.append(tuple([word,"N"]))
            
    result='\n Pos of speech : {}'.format(final_list)
    tab2_display_text.insert(tk.END,result)
    print(final_list)



# MAIN NLP TAB





# bUTTONS
button1=Button(tab1,text="Tokenize", width=12,command=get_tokens,bg='#03A9F4',fg='#fff')
button1.grid(row=4,column=0,padx=10,pady=10)


button6=Button(tab1,text="Clear Result", width=12,command=clear_display_result)
button6.grid(row=5,column=2,padx=10,pady=10)

# Display Screen For Result
tab1_display=Text(tab1)
tab1_display.grid(row=7,column=0, columnspan=3,padx=5,pady=5)

# Allows you to edit
tab1_display.config(state=NORMAL)




# FILE READING  AND PROCESSING TAB
l1=Label(tab2,text="Open File To Process")
l1.grid(row=1,column=1)


displayed_file=ScrolledText(tab2,height=7)# Initial was Text(tab2)
displayed_file.grid(row=2,column=0, columnspan=3,padx=5,pady=3)


# BUTTONS FOR SECOND TAB/FILE READING TAB
b0=Button(tab2,text="Open File", width=12,command=openfiles,bg='red')
b0.grid(row=1,column=0,padx=5,pady=3)

b2=Button(tab2,text="Tokenize", width=12,command=get_file_tokens,bg='#03A9F4',fg='blue')
b2.grid(row=3,column=0,padx=5,pady=5)



b6=Button(tab2,text="Clear Result", width=12,command=clear_result)
b6.grid(row=3,column=1,padx=2,pady=2)

b6=Button(tab2,text="pos tag ", width=12,command=get_file_pos_tags)
b6.grid(row=3,column=2,padx=6,pady=6)

tab2_display_text=ScrolledText(tab2,height=10)
tab2_display_text.grid(row=7,column=0, columnspan=3,padx=5,pady=5)


tab2_display_text.config(state=NORMAL)

window.mainloop()
```
